```python
# read csv 
import pandas as pd
import requests

top_twitter_accounts = pd.read_csv(r"C:\Users\johna\anaconda3\envs\twitter-influence-env\twitter-influence\data\01_raw\top_twitter_accounts.csv")
top_twitter_accounts.head()
```

```python
usernames = list(top_twitter_accounts['Username_API_search'])
```

```python
# Get user tweets
responses = []
formatted_data = []
#usernames = list(usernames[2])
for username in usernames:
    print(username)
    querystring = {"username":username,"limit":"100","include_replies":"false"}
    
    url = "https://twitter154.p.rapidapi.com/user/tweets"

    headers = {
        "X-RapidAPI-Key": "154b04f735msh95207f233e150b4p160e64jsnbede5ac2c0cc",
        "X-RapidAPI-Host": "twitter154.p.rapidapi.com"
    }    
        
    response = requests.request("GET", url, headers=headers, params=querystring)
    data = response.json().get('results')
    
    for entry in data:
        formatted_entry = {
            "tweet_id": entry["tweet_id"],
            "creation_date": entry["creation_date"],
            "text": entry["text"],
            "username": entry["user"]["username"],
            "name": entry["user"]["name"],
            "follower_count": entry["user"]["follower_count"],
            "following_count": entry["user"]["following_count"],
            "number_of_tweets": entry["user"]["number_of_tweets"],
            "is_verified": entry["user"]["is_verified"],
            "is_blue_verified": entry["user"]["is_blue_verified"],
            "location": entry["user"]["location"],
            'favorite_count': entry["favorite_count"],
            'retweet_count': entry["retweet_count"],
            'quote_count': entry["quote_count"],
            'retweet': entry["retweet"],
            'video_view_count': entry["video_view_count"],
            'in_reply_to_status_id': entry["in_reply_to_status_id"],
            'quoted_status_id': entry["quoted_status_id"]

        }
        formatted_data.append(formatted_entry)
```

```python
tweetdata = pd.DataFrame(formatted_data)
tweetdata
```
